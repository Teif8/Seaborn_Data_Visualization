#### Data Visualization With Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',size='size')
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
## Bar Plot
sns.barplot(x='day',y='total_bill',hue='time',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
sns.barplot(tips['day'])
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
## Box Plot
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
sns.boxplot(tips['total_bill'])
```

```python
### Histograms
sns.histplot(tips['total_bill'],kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

 HeatMap

```python
correlation = tips.corr(numeric_only=True)
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap='Blues')
```

```python
# two categorical data
import pandas as pd
cross_tab = pd.crosstab(tips['sex'], tips['smoker'])

# Heatmap visualization
sns.heatmap(cross_tab, annot=True, cmap='Blues')
```

```python
# Pairplot
sns.pairplot(tips)
```
